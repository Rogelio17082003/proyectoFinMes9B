```python
import pandas as pd
df = pd.read_csv('datos_estudiantes.csv')
df
```

```python
df.info()
```

```python
x = df.drop(['Paso'], axis=1)
x.head(4)
```

```python
y = df['Paso']
y.head(20)
```

# convertir categorico a numerico 

```python
#Letra a numero
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
categorical_columns = ['Carrera Escogida', 'Escuela de Origen',
                       'Municipio', 'Estado', 'Sexo', 'Estado Civil','calFin']
x[categorical_columns] = encoder.fit_transform(x[categorical_columns])
x
```

```python
x.info()
```

# se escalan los valores

```python
#escalar
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scalerX = scaler.fit_transform(x)
scaler_titanic = pd.DataFrame(scalerX, columns=x.columns)
scaler_titanic.head(20)


```

# seleccion de caracteristicas

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression


#Crear modelo 
modelo = LogisticRegression(solver='lbfgs', max_iter=5000) # uaumentar maximo de iter 5000 por 1000

rfe = RFE(estimator=modelo, n_features_to_select=7, step=1)
rfe = rfe.fit(scaler_titanic, y)

for feature, selected in zip(scaler_titanic.columns, rfe.support_):
    print(f'{feature}: {selected}')

# Mostrar la clasificación de las características
print('Ranking de características:', rfe.ranking_)
# Mostrar las características seleccionadas
print('Características seleccionadas:', rfe.support_)
```

```python
x_selected = scaler_titanic[scaler_titanic.columns[rfe.support_]]
x_selected.head(5)
```

```python
x_selected.info()
```

## modelo Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
rf = RandomForestClassifier(n_estimators=1000, random_state=42)

svm_scores = cross_val_score(rf, x_selected, y, cv=5, scoring='accuracy')
svm_scores.mean()
```

```python
rf_model = RandomForestClassifier(n_estimators=1000, random_state=42)
rf_model.fit(x_selected, y)
```

```python
svm_scoress = cross_val_score(rf_model, x_selected, y, cv=5, scoring='accuracy')
svm_scoress.mean()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_selected, y, test_size=0.2, random_state=42)

# Entrenar el modelo Random Forest

svm_scoress = cross_val_score(rf_model, X_test, y_test, cv=5, scoring='accuracy')
svm_scoress.mean()
```
